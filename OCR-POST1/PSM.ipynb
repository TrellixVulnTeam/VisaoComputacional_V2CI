<h3 style="text-align:center">Exemplo 1</h3>

Iremos utilizar o matplotlib para que possamos ver as imagens na mesma celula de código

```python
image_path = r"D:\Arquivos Pessoais\Faculdade\Python\Tesseract\imagens\trecho-livro.jpg"
imagem = cv2.imread(image_path)
rgb = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB) #é necessário converter de BGR(formato de cores que o OpenCV utiliza ) para RGB(Formato de cores que o matplotlib utiliza)
plt.imshow(rgb) #Utilizaremos o método imshow para visualizarmos a imagem 
plt.axis("off"); #tiramos as marcações dos eixos.
```

Iremos Utilizar o método image_to_string mas deixaremos o PSM padrão, o padrão do tesseract é o PSM 3 que é "Segmentação totalmanete automatica, mas sem OSD(Orientation and Script detection)",
E na célula seguinte usaremos o PSM-6 "Considere um unico bloco de texto uniforme". Para que possamos comparar as diferenças.

```python
print(ocr.image_to_string(rgb))
```

```python
#para que seja possivel a configuração do psm iremos adicionar a flag config dentro do método image_to_string
config_ocr = "--psm 6"
print(ocr.image_to_string(rgb, config=config_ocr))
```

Podemos analisar que tivemos diferença entre as duas células, mostrando que uma boa configuração também irá nós possibilitar maior assertividade sobre os processamentos.
